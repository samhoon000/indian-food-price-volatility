```python
import kagglehub
path = kagglehub.dataset_download("arjunyadav99/indian-agricultural-mandi-prices-20232025")
```

```python
import pandas as pd
import os
os.listdir(path)
csv_file_name = 'Agriculture_price_dataset.csv'
full_csv_path = os.path.join(path, csv_file_name)
df = pd.read_csv(full_csv_path, encoding='latin1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df['Price Date']=pd.to_datetime(df['Price Date']) #Convert the data type inot date_time
```

```python
df = df.sort_values(['Commodity', 'Price Date'])
```

```python
df['Daily_Price_Change']=df.groupby('Commodity')['Modal_Price'].diff() #Price change in the market for each commodity
```

```python
df.head()
```

```python
df['Prcnt_Price_Change']=df.groupby('Commodity')['Modal_Price'].pct_change()*100 #Percent of the price change
```

```python
df.head()
```

```python
df['Price_Spread']=df['Max_Price'] - df['Min_Price'] #Takes difference between highest and lowest price of the day
```

```python
df.head()
```

```python
df['Rolling_7D_Volatility']=df.groupby('Commodity')['Modal_Price'].rolling(7).std().reset_index(level=0,drop=True)
#How much did prices swing this week?
#High value = prices fluctuated a lot this week
#Low value = stable prices
```

```python
df['Month']=df['Price Date'].dt.month
```

```python
def get_season(month):
  if month in [12,1,2]:
    return "Winter"
  elif month in [3,4,5]:
    return "Summer"
  elif month in [6,7,8]:
    return "Monsoon"
  else:
    return "Post-Monsoon"
```

```python
df['Season']=df['Month'].apply(get_season)
```

```python
df.head()
```

```python
# Market Average price helps us identify which market charges more or less (transofrm keeps all the row in commodity)
df['Market_Average_Price']=df.groupby(['Market Name','Commodity'])['Modal_Price'].transform('mean')
```

```python
df['Price_Deviation']=df['Modal_Price'] - df['Market_Average_Price'] #Positive value = Overpriced Today and Negative value = Cheaper than usual
```

```python
df.head()
```

```python
df.to_csv("Indian_Agriculutre_Mandi_Price.csv",index=False)
```
